```python
#data = pd.read_csv('OS_survival_results.txt', sep="\t", index_col=0)

paper_match_list = ours_match_list = ['Breast']

for i in range(len(ours_match_list)):
    cancer_type = paper_match_list[i]
    prime_type = ours_match_list[i]
    dfx = pd.read_excel(gene_xlsx_file, header=1, usecols='A:E', sheet_name=cancer_type)
    
    dfx['pvalue2'] = data[prime_type+'_pvalue']
    dfx['HR2'] = data[prime_type+'_HR']
    dfx['CI_low2'] = data[prime_type+'_CI_low']
    dfx['CI_high2'] = data[prime_type+'_CI_high']

    dfx['pvalue_diff'] = (dfx['pvalue2']-dfx['pvalue']).abs()
    dfx['HR_diff'] = (dfx['HR2']-dfx['HR']).abs()
    dfx['CI_low_diff'] = (dfx['CI_low2']-dfx['CI_low']).abs()
    dfx['CI_high_diff'] = (dfx['CI_high2']-dfx['CI_high']).abs()

    dfx['pvalue_diff_percent'] = dfx['pvalue_diff']/dfx['pvalue']
    dfx['HR_diff_percent'] = dfx['HR_diff']/dfx['HR']
    dfx['CI_low_diff_percent'] = dfx['CI_low_diff']/dfx['CI_low']
    dfx['CI_high_diff_percent'] = dfx['pvalue_diff']/dfx['CI_high']

    print(dfx[dfx['pvalue_diff_percent']<0.1].shape[0])

    # plt.hist(dfx['HR_diff_percent'])

    writer = pd.ExcelWriter(output_xlsx_file, engine='xlsxwriter')
    dfx.to_excel(writer, sheet_name=cancer_type)
    writer.save()
```
