```python
# https://medium.com/analytics-vidhya/write-your-own-custom-data-generator-for-tensorflow-keras-1252b64e41c3
import tensorflow as tf

class CustomDataGen(tf.keras.utils.Sequence):
    
    def __init__(self, df, X_col, y_col,
                 batch_size,
                 s3_client,
                 s3_bucket,
                 s3_prefix,
                 input_size=(1, 2000, 1), #[1, input_len, input_channel]
                 preprocess=preprocess,
                 shuffle=True):
        
        self.df = df.copy()
        self.X_col = X_col
        self.y_col = y_col
        self.batch_size = batch_size
        self.s3_client= s3_client
        self.s3_bucket = s3_bucket
        self.s3_prefix = s3_prefix
        self.input_size = input_size
        self.preprocess = preprocess
        self.shuffle = shuffle
        self.n_classes = df[y_col].nunique()
        self.n = len(self.df)
    
    def on_epoch_end(self):
        if self.shuffle:
            self.df = self.df.sample(frac=1).reset_index(drop=True)
    
    def __get_input(self, file_path):
    
         # load file from s3
          f = self.s3_client.laod_file(self.s3_bucket,self.s3_prefix,file_path)
          # preprocess
          f = self.preprocess(f)
          return f
    
    def __get_output(self, label, num_classes):
        return tf.keras.utils.to_categorical(label, num_classes=num_classes)
    
    def __get_data(self, batches):
        # Generates data containing batch_size samples

        path_batch = batches[self.X_col]
        label_batch = batches[self.y_col]

        X_batch_start = np.asarray([self.__get_input(x)[:self.input_size[1]] for x in path_batch])
        X_batch_end = np.asarray([self.__get_input(x)[-self.input_size[1]:] for x in path_batch])

        y_batch = np.asarray([self.__get_output(y, self.n_classes) for y in label_batch])

        return tuple(X_batch_start,X_batch_end), y_batch
    
    def __getitem__(self, index):
        
        batches = self.df[index * self.batch_size:(index + 1) * self.batch_size]
        X, y = self.__get_data(batches)        
        return X, y
    
    def __len__(self):
        return self.n // self.batch_size
```
